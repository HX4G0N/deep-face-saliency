```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
import scipy.interpolate as interp
%matplotlib inline
```

```python
loss_hist = []
for i in range(23):
    loss_hist.append(pickle.load(open('save_segnet_acc_loss_' + str(i) + '.pkl', 'rb')))
```

```python
loss = []
acc_train = []
acc_val = []
val_loss = []
for i in range(len(loss_hist)):
    loss.append(loss_hist[i]['train_loss_history'])
    acc_train.append(loss_hist[i]['train_acc_history'])
    acc_val.append(loss_hist[i]['val_acc_history'])
    val_loss.append(loss_hist[i]['val_loss_history'])
```

```python
loss = np.hstack(loss)
acc_train = np.hstack(acc_train)
acc_val = np.hstack(acc_val)
val_loss = np.hstack(val_loss)
```

```python
val_loss_stretch = interp.interp1d(np.arange(val_loss.size),val_loss)
val_loss_stretch = val_loss_stretch(np.linspace(0,val_loss.size-1,loss.size))
```

```python
plt.rcParams.update({'font.size': 12, 'font.family': 'sans'})
fig = plt.figure()
ax = fig.add_axes([0.2, 0.2, 0.8, 0.8])
ax2 = fig.add_axes([0.7, 0.7, 0.3, 0.3])
ax.plot(loss)
ax2.plot(val_loss_stretch, 'r')
ax.set_title('Loss trend SegNet')
ax.set_xlabel('# of iterations')
ax.set_xlim([0, 5000])
ax.legend(['Training loss'], loc=3)
ax2.legend(['Val loss'])
fig.savefig("loss_seg.png", dpi=300, bbox_inches = 'tight')
```
